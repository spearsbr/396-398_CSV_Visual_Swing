```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Test_Data.csv") # load the CSV file
df.head() # Displays the CSV files
```

```python
df['t'] = (df['ms'] - df['ms'].iloc[0]) / 1000.0
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['t'], df['ax'], label='ax')
plt.plot(df['t'], df['ay'], label='ay')
plt.plot(df['t'], df['az'], label='az')
plt.title("Accelerometer vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['t'], df['gx'], label='gx')
plt.plot(df['t'], df['gy'], label='gy')
plt.plot(df['t'], df['gz'], label='gz')
plt.title("Gyroscope vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Angular Velocity (rad/s)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['t'], df['roll'], label='roll')
plt.plot(df['t'], df['pitch'], label='pitch')
plt.plot(df['t'], df['yaw'], label='yaw')
plt.title("Orientation (RPY) vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Angle (deg)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(df['t'], df['qw'], label='qw')
plt.plot(df['t'], df['qx'], label='qx')
plt.plot(df['t'], df['qy'], label='qy')
plt.plot(df['t'], df['qz'], label='qz')
plt.title("Quaternion Components vs Time")
plt.xlabel("Time (s)")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

```python
gyro = df[['gx','gy','gz']].values.astype(float)
gyro_mag = np.linalg.norm(gyro, axis=1)

plt.figure(figsize=(10,4))
plt.plot(df['t'], gyro_mag, color='purple')
plt.title("Overall Swing Motion (Gyro Magnitude)")
plt.xlabel("Time (s)")
plt.ylabel("Angular Speed (rad/s)")
plt.grid(True)
plt.show()
```
